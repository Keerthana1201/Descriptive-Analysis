```python
import pandas as pd

df = pd.read_csv("../../Dataset.csv")
```

```python
num_df = df.select_dtypes(include=['int64', 'float64'])
stats = num_df.describe().T
stats.to_csv('level1_task2_numerical_stats.csv')
stats
```

```python
for col in ['Country Code', 'City', 'Cuisines']:
    if col in df.columns:
        vc = df[col].value_counts()
        vc.to_csv(f'level1_task2_{col.replace(" ", "_")}_value_counts.csv', header=[col])
```

```python
if 'Cuisines' in df.columns:
    cuisines_series = df['Cuisines'].dropna().astype(str).str.split(',')
    cuisines_exploded = cuisines_series.explode().str.strip()
    top_cuisines = cuisines_exploded.value_counts().head(50)
    top_cuisines.to_csv('level1_task2_top_cuisines.csv')
```

```python
target_candidates = [c for c in df.columns if 'rating' in c.lower() or 'aggregate' in c.lower()]
target_col = None
for c in target_candidates:
    if c.lower().replace(' ', '') in ("aggregaterating", "aggregate_rating"):
        target_col = c
        break
if not target_col and target_candidates:
    target_col = target_candidates[0]

if target_col:
    df[target_col] = pd.to_numeric(df[target_col], errors='coerce')
    dist = df[target_col].value_counts(dropna=False).sort_index()
    dist.to_csv('level1_task2_target_distribution.csv')

```
